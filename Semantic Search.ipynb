```python
from odt.opendatasemanticframework import OpenDataSemanticFramework
filename = 'odt-ontology-v1.0.rdf'
ontology = OpenDataSemanticFramework(filename)
ccs = ontology.compute_ccs()
##
ontology.register_dataset('anleggsregister.rdf')
ontology.register_dataset('http://78.91.98.234:5000/dataset/trafikkdata-fra-jernbaneverket.rdf')
```

```python
##
ontology.register_similarity('anlegg.rdf')
ontology.register_similarity('trafikk-jernbane.rdf')
```

```python

query = "buildings with weather report of road"
xs = ontology.search_query(query)
for x, score in xs:
    print (ontology.get_dataset_info(x), x, score)
```
